```python
print("Dear Customer, Welcome to our shop!")

name = input("Please enter your name: ")
address = input("Please enter your address: ")
distance = float(input("Distance from home to shop: "))

#Menu of the available items
print("Below is the menu for your shopping:")
print(" ________________________________________")
print("|      #ESSENTIAL ITEMS AVAILABLE#       |")
print("|","%5s %10s %10s %10s"%("S.NO","Items","Quantity","Cost/Item"),"|")
print("|","%5d %10s %10d %10.2f"%(1,"Biscuits",5,20.50),"|")
print("|","%5d %10s %10d %10.2f"%(2,"Cereals",10,90.00),"|")
print("|","%5d %10s %10d %10.2f"%(3,"Chicken",20,200.00),"|")
print("|","%5d %10s %10d %10.2f"%(4,"Oats",20,39.99),"|")
print("|","%5d %10s %10d %10.2f"%(5,"Rice",30,79.99),"|")
print("|________________________________________|")

item_details = {'biscuits': [5,20.50],
                'cereals': [10,90],
                'chicken': [20,200],
                'oats': [20,39.99],
                'rice': [30,79.99]}

#function to calculate item cost along with purchase quantity
cost = []                                                                       #Storing the cost value of each purchase in a list
def item_cost(i_cost):
  i_cost = quant * item_details[ask][1]                                         #item_details[ask][1]:- this is to get the second value from the dictionary
  cost.append(i_cost)

#function to ask customer for continue shopping or not
def ask_for_shop():
  check = input("Do you want to continue shopping? y/n: ").lower()
  if check == 'n' or check == 'no':
    return False

#function which calculates delivery charge based on the distance from home
def del_charge():
  if 0< distance <= 5:
    charge = 20
  elif 5< distance <= 10:
    charge = 40
  else:
    charge = 50
  return charge

#for storing the customer input of shopping items
item_names = []
item_price = []
item_quantity = []
count = 1
while True:
  ask = input("Please enter the item you want to buy: ").lower()
  if ask in item_details.keys():
    quant = int(input(f"How many {ask} packets you require?  "))   
    if quant > item_details[ask][0]:                                            #Condition to check entered quantity exceeds the limit or not
      print("Sorry! Item quantity exceeded. Out of Stock!!!")
      if ask_for_shop() == False:                                               #ask_for_shop(): User defined function(checks whether customers wants to proceed with the shopping)
        break
    else:
      item_cost(quant)                                                          #item_cost(): User defined function. returns quantity*item cost
      item_names.append(ask)                                                    #Stores item names taken from the user to a list
      item_price.append(item_details[ask][1])                                   #Stores price of the item w.r.t name
      item_quantity.append(quant)                                               #Takes quantity of the item to the blank list taken from the user
      count += 1
      if ask_for_shop() == False:
        break
  else:
    print("Oops! Entered Item not found.")
    if ask_for_shop() == False:
      break

#for getting bill with item details
print("\n")
print(" ________________________________________")
print("|                 BILL                   |")
print("|","%5s %10s %10s %10s"%("S.NO","Items","Quantity","Total_Cost"),"|")
i=1
while i in range(count):                                                        #This i is to get SL.NO in the bill
  for j in range(count-1):                                                      #here j starts from 0th posistion and count starts from 1. for getting the list values
    print("|","%5d %10s %10d %10.2f"%(i, item_names[j], item_quantity[j], item_quantity[j] * item_price[j]),"|")
    i += 1
    j += 1
print("|________________________________________|")

#Final Bill calculation
total_item_cost = sum(cost)
total_bill_cost = total_item_cost + del_charge()                                #del_charge():- user defined function

#Final Bill amount in the output
print("Total Item Cost  : ", total_item_cost)
print("Delivery Charge  : ", del_charge())
print("Total Bill Amount:  Total Item Cost + Delivery Charge: ", total_bill_cost)
print("\n")
print("***Customer Details***")
print("Customer Name: ", name)
print("Address      : ", address)
print("Distance     : ", distance)
print("\n\n")
print(f"Thank you for shopping {name}! Have a great day!")
```
